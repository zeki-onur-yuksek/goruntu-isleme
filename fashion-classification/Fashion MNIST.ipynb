# STEP #4: TRAINING THE MODEL

```python
X_train = training[:,1:]/255
y_train = training[:,0]

X_test = testing[:,1:]/255
y_test = testing[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_validate, y_train, y_validate = train_test_split(X_train, y_train, test_size = 0.2, random_state = 12345)
```

```python
X_train = X_train.reshape(X_train.shape[0], *(28, 28, 1))
X_test = X_test.reshape(X_test.shape[0], *(28, 28, 1))
X_validate = X_validate.reshape(X_validate.shape[0], *(28, 28, 1))
```

```python
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout, Input
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Eğer X_train, y_train direkt verildiyse eğitim/doğrulama ayırımı:
X_train, X_validate, y_train, y_validate = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)
```

```python
cnn_model = Sequential()
cnn_model.add(Input(shape=(28, 28, 1)))  # Giriş katmanı
cnn_model.add(Conv2D(64, (3, 3), activation='relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))
cnn_model.add(Dropout(0.25))

# İkinci konvolüsyon katmanı (isteğe bağlı)
# cnn_model.add(Conv2D(32, (3, 3), activation='relu'))
# cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

# Dense katmanlar
cnn_model.add(Flatten())
cnn_model.add(Dense(units=32, activation='relu'))
cnn_model.add(Dense(units=10, activation='softmax'))  # Çok sınıflı sınıflandırma için
```

```python
cnn_model.compile(
    loss='categorical_crossentropy',  # one-hot encode edilmiş etiketlere uygun
    optimizer=Adam(learning_rate=0.001),
    metrics=['accuracy']
)
```

```python
epochs = 5

history = cnn_model.fit(
    X_train,
    y_train,
    batch_size=128,
    epochs=epochs,
    verbose=1,
    validation_data=(X_validate, y_validate)
)
```

# STEP #5: EVALUATING THE MODEL

```python
evaluation = cnn_model.evaluate(X_test, y_test)
print('Test Accuracy : {:.3f}'.format(evaluation[1]))
```

```python
# Tahmin yap
predictions = cnn_model.predict(X_test)

# En yüksek olasılığa sahip sınıfı al (one-hot olduğundan)
predicted_classes = np.argmax(predictions, axis=1)

# Gerçek sınıfları da düzle (çünkü one-hot)
true_classes = np.argmax(y_test, axis=1)
```

```python
import matplotlib.pyplot as plt

L = 5
W = 5
fig, axes = plt.subplots(L, W, figsize=(12, 12))
axes = axes.ravel()

for i in np.arange(0, L * W):
    axes[i].imshow(X_test[i].reshape(28, 28), cmap='gray')
    axes[i].set_title("Prediction Class = {}\nTrue Class = {}".format(
        predicted_classes[i],
        np.argmax(y_test[i])
    ))
    axes[i].axis('off')

plt.subplots_adjust(wspace=0.5)
plt.tight_layout()
plt.show()
```
